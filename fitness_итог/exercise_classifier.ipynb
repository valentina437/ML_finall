```python
import os
import glob
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

class MLClassifier:
    pass  
```

```python
def load_gif_files(self):
    gif_files = {}
    for folder in self.gif_folders.values():
        gifs = glob.glob(os.path.join(folder, "*.gif"))
        gif_files[os.path.basename(folder)] = gifs
    return gif_files

MLClassifier.load_gif_files = load_gif_files
```

```python
def __init__(self):
    self.vectorizer = TfidfVectorizer(max_features=300)
    self.model = LogisticRegression(max_iter=1000)
    self.label_encoder = LabelEncoder()
    self.is_trained = False
    
    self.gif_folders = {
        'Руки': r"C:\Users\user\Desktop\fitness_итог\hands",
        'Спина': r"C:\Users\user\Desktop\fitness_итог\back", 
        'Пресс': r"C:\Users\user\Desktop\fitness_итог\press",
        'Грудь': r"C:\Users\user\Desktop\fitness_итог\breast",
        'Ноги': r"C:\Users\user\Desktop\fitness_итог\legs",
        'Плечи': r"C:\Users\user\Desktop\fitness_итог\shoulders"  
    }
    
    self.gif_files = self.load_gif_files()

MLClassifier.__init__ = __init__
```

```python
def train(self):
    exercises = ['жим лежа', 'приседания', 'подтягивания', 'бицепс', 'плечи', 'пресс']
    muscle_groups = ['Грудь', 'Ноги', 'Спина', 'Руки', 'Плечи', 'Пресс']
    
    X = self.vectorizer.fit_transform(exercises)
    y = self.label_encoder.fit_transform(muscle_groups)
    self.model.fit(X, y)
    self.is_trained = True
    return True

MLClassifier.train = train
```

```python
def predict_simple(self, query):
    q = query.lower()
    if 'груд' in q or 'жим' in q: return 'Грудь'
    if 'спин' in q or 'тяга' in q: return 'Спина'
    if 'ног' in q or 'присед' in q: return 'Ноги'
    if 'бицепс' in q or 'трицепс' in q or 'руки' in q: return 'Руки'
    if 'плеч' in q or 'дельт' in q: return 'Плечи'
    if 'пресс' in q or 'живот' in q: return 'Пресс'
    return 'Грудь'

MLClassifier.predict_simple = predict_simple
```

```python
def predict(self, query):
    if self.is_trained:
        vec = self.vectorizer.transform([query.lower()])
        idx = self.model.predict(vec)[0]
        return self.label_encoder.inverse_transform([idx])[0]
    return self.predict_simple(query)

MLClassifier.predict = predict
```

```python
def get_gif_for_exercise(self, exercise_name):
    muscle_group = self.predict(exercise_name)
    if muscle_group in self.gif_files and self.gif_files[muscle_group]:
        import random
        return random.choice(self.gif_files[muscle_group])
    return None

MLClassifier.get_gif_for_exercise = get_gif_for_exercise
```
